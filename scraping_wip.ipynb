```python
# QUERY PARAMETERS
# Date format: MMDDYYYY
beginDate = '02092021'
endDate = '02102021'
contentType = 'csv'  # csv, html
params = {
    'beginDate': beginDate,
    'endDate': endDate,
    'contentType': contentType
}
```

```python
with requests.Session() as s:
    raw_response = s.get(config_scraper.host + config_scraper.filename, params=params)
    # save raw response to /raw_reports

    # prepare a pandas object for each date and save it to pd_ready_reports
    if contentType == 'csv':
        dates, closing_prices, core_data = scraper.process_csv(raw_response)
        print('dates', dates)
        print('closing', closing_prices)
        print('core', core_data)
    elif contentType == 'html':
        print('html contentType currently not supported')
```
