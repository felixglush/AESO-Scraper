```python
import numpy as np
import pandas as pd
```

```python
index = ['Pool Participant ID', ' Asset Type', ' Asset ID']
hours_row = ['Hour 1', 'Hour 2', 'Hour 3', 'Hour 4', 'Hour 5', 'Hour 6', 'Hour 7', 'Hour 8', 'Hour 9', 'Hour 10', 'Hour 11', 'Hour 12', 'Hour 13', 'Hour 14', 'Hour 15', 'Hour 16', 'Hour 17', 'Hour 18', 'Hour 19', 'Hour 20', 'Hour 21', 'Hour 22', 'Hour 23', 'Hour 24']
hour_values = ['45.45', '48.64', '47.34', '44.89', '43.44', '70.49', '218.73', '438.24', '558.40', '552.43', '580.21', '506.76', '377.32', '168.27', '289.80', '325.83', '499.20', '627.39', '567.60', '484.54', '94.33', '65.78', '46.82', '44.73']

dates = ['May 10, 2021']

data_index = [['9496', 'RETAILER', '941A'], 
              ['9496', 'RETAILER', '941B']]

# generates 2 rows of random number data (for 24 hours) for each date
data_core = np.random.rand(2, 24) 
```

```python
hours_core = pd.DataFrame(data=data_core, columns=hours_row)
hours_core
```

```python
index_df = pd.DataFrame(data=data_index, columns=index)
index_df
```

```python
concatenated = pd.concat([index_df, hours_core], axis=1).set_index(index)
concatenated
```

```python
# concatenated.columns = pd.MultiIndex.from_tuples(zip(hours_core.columns, hour_values))
concatenated
```

```python
concatenated_pivot = concatenated.stack(0).reset_index()
concatenated_pivot.columns = ['Pool Participant ID', 'Asset Type', 'Asset ID', 'Hour', 'MWh']
concatenated_pivot
```

```python
target = concatenated_pivot.drop(['Pool Participant ID','Asset Type'], axis=1)
target['Hour'] = target.apply(lambda row: row.Hour.split(' ')[1], axis=1).astype(int)
target['Peak Status'] = np.where((target.Hour >= 7) & (target.Hour <= 10), 'Peak', 'Off-Peak')
target
```

```python
concatenated_reset_index = concatenated.reset_index()
concatenated_reset_index
```

```python
concatenated_reset_index.to_csv('test.csv')
```
